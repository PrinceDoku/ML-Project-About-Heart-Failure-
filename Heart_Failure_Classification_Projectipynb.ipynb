```python
# loaidng and importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# Loading the Heart failure Dataset
df =pd.read_csv('/content/heart_failure_clinical_records_dataset (1).csv')
df
```

```python
# Viewing the head of the Dataset
df.head()
```

```python
# Checking for null values (No null values reported)
df.isnull().sum()
```

```python
# summary of the dataset
df.describe()
```

```python
# Divide the dataset into X and Y(target)
X = df.drop(columns='DEATH_EVENT', axis=1)
Y = df['DEATH_EVENT']
```

```python
X
```

```python
Y
```

```python
# Check the value counts of Y(target)
Y.value_counts()
```

```python
# Checking the standard deviation of the Inputs
X.std()
```

```python
# Standardize the inputs (X)
scaler = StandardScaler()
Scaled_X = scaler.fit_transform(X)
```

```python
# Viewing of the scaled data
Scaled_X
```

```python
Scaled_X.std()
```

```python
## The std of the inputs (X) is equal to 1 so our inputs are scaled well
```

There is no need to label encode our Targets (Death_event) since they are already encoded in 0 & 1

```python
# Splitting our dataset into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(Scaled_X, Y, test_size=0.25, stratify=Y, random_state=42)
```

```python
X_train
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
Y_train
```

**bold text** LOGISTIC REGRESSION MODEL AS THE FIRST MODEL

```python
model_1 =LogisticRegression()
```

```python
model_1, model_1.fit(X_train, Y_train)
```

```python
model_1_predictions= model_1.predict(X_test)
```

```python
score = accuracy_score(model_1_predictions, Y_test)
print('The score of the first model is:', score)
```

The Accuracy_score of the First model(Logistic regression) for testing is 83%

```python
score_training = accuracy_score(model_1.predict(X_train), Y_train)
print('The score of the first model on testing data is:', score_training)
```

The Accuracy_score of the First model(Logistic regression) for the training is 85%

```python
# Using sample data from the dataset to predict the death_event
input = [	62.0,0,61,1,38,1,155000.00,1.1,143,1,1,270]
input_array = np.asarray(input)
input_reshaped = input_array.reshape(1,-1)
std_data = scaler.transform(input_reshaped)
prediction = model_1.predict(std_data)
print(prediction)
```

**bold text** SUPPORT VECTOR MACHINE AS SECOND MODEL

```python
model_2 = SVC()
```

```python
model_2.fit(X_train, Y_train)
```

```python
model_2_predictions = model_2.predict(X_test)
```

```python
score = accuracy_score(model_2_predictions, Y_test)
print('The score of the second model is:', score)
```

The Accuracy_score of the second model(Support vector machine) for testing is 79%

```python
score_training = accuracy_score(model_2.predict(X_train), Y_train)
print('The score of the second model on testing data is:', score_training)
```

The Accuracy_score of the Second model(Logistic regression) for the training is 90%

```python
# Using sample data from the dataset to predict the death_event
input = [	62.0,0,61,1,38,1,155000.00,1.1,143,1,1,270]
input_array = np.asarray(input)
input_reshaped = input_array.reshape(1,-1)
std_data = scaler.transform(input_reshaped)
prediction = model_2.predict(std_data)
print(prediction)
```

From the two models used in training the Heart failure dataset, Logistic regression came out as the best model
